```python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
```

```python
# ðŸ”¥ Overdetermined ðŸ”¥
# I have more equations than variables
# equations > vars

x = np.arange(-6, 6) #This is the interval where the values are taken

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x + 1

plt.figure() #For each of them, let's create the figure / graph

plt.plot(x, y_1, color="red") #
plt.plot(x, y_2, color="blue")
plt.plot(x, y_3, color="green")

plt.xlim(-3, 8) #The numbers below the chart
plt.ylim(-4, 8) #If you move the chart, these numbers don't work anymore

plt.axvline(x=0, color="black") # The Vertical line
plt.axhline(y=0, color="black")

plt.show()
plt.savefig("photosaved.png")

# ðŸ‘†
#ðŸ”´ The three lines should be colliding. But they aren't. 
#ðŸ”´ This is a Overdetermined system: we have more equations than variables,
#ðŸ”´ So it has no solution.
```

```python
# ðŸ”¥ Determined ðŸ”¥
# Same equations and variables. 2 and 2 in this case

x = np.arange(-6, 6) #This is the interval where the values are taken

y_2 = -x+3
y_3 = 2*x + 1

plt.figure() #For each of them, let's create the figure / graph

plt.plot(x, y_2, color="blue")
plt.plot(x, y_3, color="green")

plt.xlim(-3, 8) #The numbers below the chart
plt.ylim(-4, 8) #If you move the chart, these numbers don't work anymore

plt.axvline(x=0, color="black") # The Vertical line
plt.axhline(y=0, color="black")

plt.show()
```

```python
# ðŸ”¥ Undetermined ðŸ”¥
# There's nothing between a solution and infinite solutions
# We have systems with 0, with 1, or with infinite

x = np.arange(-6, 6) 

y_3 = 2*x + 1

plt.figure() #For each of them, let's create the figure / graph

plt.plot(x, y_3, color="green")

plt.xlim(-3, 8) #The numbers below the chart
plt.ylim(-4, 8) #If you move the chart, these numbers don't work anymore

plt.axvline(x=0, color="black") # The Vertical line
plt.axhline(y=0, color="black")

plt.show()
```
